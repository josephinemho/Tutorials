## Import and clean data

```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.externals import joblib
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('../work/data/device_failure.csv', engine='python')
```

```python
data.head(2)
```

### Do some general checks

```python
# Check for nulls

data.isnull().sum()
```

```python
# Check datatypes

for col in data:
    print(col, "\t", data[col].dtype)
```

### Handle `['date']` variable
From checking the datatypes of each column, we see that `date` is an object. We need to convert it to a `datetime` object and extract more features from it.

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data['date'].dtype
```

```python
dayofweek_list = []
month_list = []
year_list = []

for i in data['date']:
    dayofweek_list.append(datetime.weekday(i))
    month_list.append(datetime.date(i).month)
    year_list.append(datetime.date(i).year)
data['Dayofweek'] = pd.Series(dayofweek_list)
data['Month'] = pd.Series(month_list)
data['Year'] = pd.Series(year_list)
```

```python
data.head(2)
```

```python
# Pretty consistent throughout the week, no major trends (so probably not triggered by humans)

plt.hist(data['Dayofweek'], bins=7, rwidth=.9)
plt.title("Frequency of failures")
plt.xlabel("Day of the week")
plt.ylabel("Frequency")
plt.show()
```

### Continue with general checks

```python
# Check the range of unique variables

for col in data:
    if len(data[col].unique()) < 10:
        print(col, data[col].unique())
    else:
        print(col, "\t", data[col].nunique())
```

```python
# We can drop ['Month'] and ['Year'] since there's only one value

data.drop(['Month', 'Year'], axis=1, inplace=True)
data.head(2)
```

```python
# Check if we're dealing with imbalanced classes

data['failure'].value_counts()
```

```python
1-(106/(124388+106))
```

Our classes are super imbalanced. So even if our model predicts there are no alerts 100% of the time, we'd get a 99% accurary score. That's not effective. We must deal with imbalanced classes before modeling.

## Transformations
Transform numeric and categorical variables so that it feeds in nicely in a way that models can understand. 

**"Garbage in, garbage out"**


Building models on top of raw data wouldn't get us the desired results or performance, as algorithms are not intelligent enough (yet) to automaticaly extract meaningful features from raw data.

```python
data.head(2)
```

```python
data.columns
```

```python
# Split nnumeric & categorical data (create "other" list just to keep track)

cat_list = ['Dayofweek']
num_list = ['attribute1', 'attribute2', 'attribute3', 'attribute4', 'attribute5', 'attribute6', 'attribute7', 'attribute8', 'attribute9']
other_list = ['date', 'failure', 'device']
```

### Deskew numeric variables
Use log transform to deskew highly skewed features. Identify which features to apply log transform to by comparing the before and after. Do not deskew features that are normally distributed or skewed to the left.

```python
num_data = data[num_list]
```

```python
# Before log transform

fig = plt.figure(figsize=(20,20))

for i, col in enumerate(num_data.columns):
    fig.add_subplot(4,3,1+i)
    sns.distplot(num_data[col], label=col)
    plt.axvline(num_data[col].mean(), c='r')
    plt.axvline(num_data[col].median(), c='g')
    plt.axvline(num_data[col].mode()[0], c='b')
    plt.legend()
```

```python
# After log transform

logged_num_data = num_data.copy()

for col in logged_num_data.columns:
    logged_num_data[col] = np.log(1+logged_num_data[col])
    
fig = plt.figure(figsize=(20,20))

for i, col in enumerate(logged_num_data.columns):
    fig.add_subplot(4,3,1+i)
    sns.distplot(logged_num_data[col], label=col)
    plt.axvline(logged_num_data[col].mean(), c='r')
    plt.axvline(logged_num_data[col].median(), c='g')
    plt.axvline(logged_num_data[col].mode()[0], c='b')
    plt.legend()
```

Merge (selected) original numeric features + log transformed features

```python
# Double-check shape and indexes before merging

print(logged_num_data.shape)
print(logged_num_data.iloc[:,:2].tail(2))
print("\t")
print(num_data.shape)
print(num_data.iloc[:,:2].tail(2))
```

```python
logged_num_data = pd.concat([logged_num_data[['attribute2', 'attribute3', 'attribute4', 'attribute5', 'attribute7', 'attribute8', 'attribute9']],num_data[['attribute1','attribute6']]], axis=1)
logged_num_data.head(2)
```

### Encode categorical variables

```python
cat_data = data[cat_list].astype('object')
```

```python
print("original size: ", cat_data.shape)
encoded_cat_data = pd.get_dummies(cat_data)
print("new size: ", encoded_cat_data.shape)
```

### Merge transformed variables

```python
print(encoded_cat_data.shape)
print(logged_num_data.shape)
print(data.shape)
```

```python
# Merge transformed categorical & numeric variables

logged_features = pd.concat([encoded_cat_data,logged_num_data], axis=1)
target = data['failure']
```

### Scale
Usually train-test-split before scale, but in the interest of time I'm just going to scale everything.

```python
data.columns
```

```python
scaler = StandardScaler()
transformed_features = scaler.fit_transform(logged_features)
```

```python
non_transformed_features = data[['attribute1', 'attribute2', 'attribute3', 'attribute4', 'attribute5', 'attribute6', 'attribute7', 'attribute8', 'attribute9', 'Dayofweek', 'device']]
```
